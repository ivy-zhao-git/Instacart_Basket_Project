### 2. Create "price_label" column.

```python
# Create a subset of first million rows
df = df_ords_prods_merge[:1000000]
```

```python
# Define a function for price_lable
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function.
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the values in new column.
df['price_range'].value_counts(dropna = False)
```

```python
# Check the price range
df['prices'].describe()
```

```python
# Apply price range using the predefined loc() Function 
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Use loc() method on entire ords_prods_merge dataframe
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

### 3. Create "busiest_day" column.

```python
# Find frequency of orders_day_of_week.
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Use for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).

result = []
for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create a new column within ords_prods_merge dataframe and set it equal to result.
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### 4. Update "busiest day" column to "busiest days" to identify two "Busiest days" and two "Slowest days" of the week.

```python
# Two busiest days (0, 1), two slowest days (4, 3), and regularly busy (all other days).
```

```python
# Create column “busiest_days" from "orders_day_of_week".

result_2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

### 5. Check values of new "busiest days" column for accuracy with observations.

```python
# Create new column 'busiest_days' within ords_prods_merge dataframe and set it equal to result_2.
df_ords_prods_merge['busiest_days'] = result_2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

The total value counts in the ords_prods_merge dataframe match the combined sum of all categories ("Regularly busy," "Busiest days," and "Slowest days"). For instance, "Busiest days" totals 11862627, which equals the sum of orders_day_of_week 0 and 1 (6203329 + 5659298).

### 6. Create a new column "busiest_period_of_day" to identify the busiest hours of the day.

```python
# Check value counts in "order_hour_of_day" column.
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

The value counts above are divided into three equal groups: "Most orders" includes 10, 11, 14, 15, 13, 12, 16, and 9; "Fewest orders" includes 23, 6, 0, 1, 5, 2, 4, and 3; "Average orders" consists of all other values in order_hour_of_day.

```python
# Use for-loop to label three periods of time as “Most orders,” “Average orders,” and “Fewest orders.”

result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Create "busiest_period_of_day" column within ords_prods_merge.
df_ords_prods_merge['busiest_period_of_day'] = result_3
```

### 7. Print frequency of "busiest_period_of_day" column.

```python
# Print value counts in "busiest_period_of_day" column.
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
