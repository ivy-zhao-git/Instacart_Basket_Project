## 4. Check the difference in expenditure (the “prices” column) depending on the hour of the day.

## 5. Determine whether there’s a connection between age and family situation by creating a line chart.

```python
# Select from the sample the columns we need
instacart_3 = small [['age','n_dependants']]
```

```python
# Check dimensions of the subset
instacart_3.head(5)
```

```python
# Create a line chart
line_2 = sns.lineplot(data = instacart_3, x = 'age',y = 'n_dependants')
```

As we see, the number of dependents fluctuates up and down with age, showing no clear correlation.

## 6. Explore whether there’s a connection between age and spending power (income) using a scatterplot.

```python
# Select from the sample the columns we need
instacart_4 = small [['age','income']]
```

```python
# Create a scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = instacart_4)
```

There is no clear trend or strong relationship between age and income; income does not consistently increase or decrease with age.

No age group showing an apparent dominance in income level, but there is a sharp rise in income levels around the age of 40.

A significant portion of data points is concentrated in the lower income range (below 200,000), regardless of age.
